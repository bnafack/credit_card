# Data preparation

```python
y_train = X_train['Class']
x_train =  X_train.drop(columns=['Class'],axis=1)

y_test = X_test['Class']
x_test =  X_test.drop(columns=['Class'],axis=1)
```

# 3) Modeling

```python
from xgboost import cv
import xgboost as xgb
params = {'colsample_bytree': 0.5,'learning_rate': 0.001,
                'max_depth': 15, 'alpha': 5}

data_dmatrix = xgb.DMatrix(data=x_train,label=y_train,enable_categorical=False)

xgb_cv_s = cv(dtrain=data_dmatrix, params=params, nfold=10,
                    num_boost_round=20, early_stopping_rounds=10, metrics="auc", as_pandas=True, seed=123)
print('The maximun auc is ',xgb_cv_s['test-auc-mean'].max())
print('The minimun auc is ',xgb_cv_s['test-auc-mean'].min())
```

```python
from sklearn.model_selection import GridSearchCV
param_test1 = {
 'max_depth':[3,5,15],
 'learning_rate': [0.1],
 'min_child_weight':[5,10]
}
gsearch1 = GridSearchCV(estimator = xgb.XGBClassifier(tree_method="gpu_hist", learning_rate =0.1, n_estimators=140,
                                                      gamma=0, subsample=0.8, colsample_bytree=0.8,
                                                      objective= 'binary:logistic', nthread=4, seed=27), 
                                                      param_grid = param_test1, scoring='roc_auc',n_jobs=4, cv=10, )
gsearch1.fit(x_train,y_train)
```

```python
gsearch1.best_params_, gsearch1.best_score_
```

```python
model = xgb.XGBClassifier(tree_method="gpu_hist", params=gsearch1.best_params_, n_estimators=140,
                          min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=1, seed=27)

model.fit(x_train, y_train)
#model.save_model("model/categorical-model.json")
```

```python
y_pred = model.predict(x_test)
y_test = np.array(y_test).astype(int)
auc_lr = roc_auc_score(y_test, y_pred)
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test, y_pred)
matthews_corrcoef(y_test, y_pred)
```

# Upsampling

In a Machine Learning problem, make sure to upsample/downsample ONLY AFTER you split into train, test (and validate if you wish). If you do upsample your dataset before you split into train and test, there is a high possibility that your model is exposed to data leakage. See an Example below.

```python
#Set the minority class to a seperate dataframe
df_1 = X_train[X_train['Class'] == 1]
#Set other classes to another dataframe
other_df = X_train[X_train['Class'] != 1]  
#Upsample the minority class
df_1_upsampled = resample(df_1,random_state=42,n_samples=len(other_df),replace=True)
#Concatenate the upsampled dataframe
df_upsampled = pd.concat([df_1_upsampled,other_df])
df_upsampled['Class'].value_counts()
```

```python
y_train_up = df_upsampled['Class']
x_train_up =  df_upsampled.drop(columns=['Class'],axis=1)

model_upsampling = xgb.XGBClassifier(tree_method="gpu_hist", params=gsearch1.best_params_, n_estimators=140,
                          min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=1, seed=27)

model_upsampling.fit(x_train_up, y_train_up)
# model_upsampling.save_model("model/upsampling-model.json")

# Prediction with the model with upsampling dataset

y_pred_up = model_upsampling.predict(x_test)
y_test_up = np.array(y_test).astype(int)

# AUC 
auc_lr_up = roc_auc_score(y_test, y_pred_up)
fpr_lr_up, tpr_lr_up, thresholds_lr_up = roc_curve(y_test, y_pred_up)
```

```python
matthews_corrcoef(y_test, y_pred_up)
```

# Downsampling

```python
#set the minority class to a seperate dataframe
df_2 = X_train[X_train['Class'] == 1]
#set majority classes to another dataframe
other_df = X_train[X_train['Class'] != 1]  
#Downsample the majority class
df_2_upsampled = resample(other_df,random_state=42,n_samples=len(df_2),replace=False) # we set replace to false to avoid data duplication
#concatenate the upsampled dataframe
df_downsampled = pd.concat([df_2_upsampled,df_2])
df_downsampled['Class'].value_counts() # check the distribution of the datatset 
```

```python
y_train_down = df_downsampled['Class']
x_train_down =  df_downsampled.drop(columns=['Class'],axis=1)
model_downsampling = xgb.XGBClassifier(tree_method="gpu_hist", params=gsearch1.best_params_, n_estimators=140,
                          min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8, scale_pos_weight=1, seed=27)

model_downsampling.fit(x_train_down, y_train_down)
# model_downsampling.save_model("model/upsampling-model.json")

# Prediction 

y_pred_down = model_downsampling.predict(x_test)
y_test = np.array(y_test).astype(int)

# AUC
auc_lr_down = roc_auc_score(y_test, y_pred_down)
fpr_lr_down, tpr_lr_down, thresholds_lr_down = roc_curve(y_test, y_pred_down)
```

```python

plt.figure(figsize=(12, 7))
plt.plot(fpr_lr, tpr_lr, label=f'AUC baseline model = {auc_lr:.2f}')
plt.plot(fpr_lr_down, tpr_lr_down, label=f'AUC downsampling = {auc_lr_down:.2f}')
plt.plot(fpr_lr_up, tpr_lr_up, label=f'AUC upsampling = {auc_lr_up:.2f}')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--', label='Baseline')
plt.title('ROC Curve', size=20)
plt.xlabel('False Positive Rate', size=14)
plt.ylabel('True Positive Rate', size=14)
plt.legend()
```
